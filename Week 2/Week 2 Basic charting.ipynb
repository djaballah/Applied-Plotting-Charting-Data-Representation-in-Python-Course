# ScatterPlots

```python
import numpy as np

x = np.array(range(8))
y = x

plt.figure()
plt.scatter(x, y, s=150, c=x)
```

```python
colors = ["green"] * (len(x) - 1)
colors.append('red')

plt.figure()
plt.scatter(x, y, color=colors)
```

```python
zip_gen = zip(list(x), list(y))
```

```python
zip_list = list(zip_gen)
zip_list
```

```python
zip_unpack = zip(*zip_list)
x, y = list(zip_unpack)
```

```python
plt.figure()

plt.scatter(x[:2], y[:2], s=50, c='green', label='City')
plt.scatter(x[2:], y[2:], s=150, c='red', label='United')
```

```python
plt.xlabel('Num Of Trophies')
plt.ylabel('Num OF Signings')
plt.title('Number of trophies per signing')
plt.legend(loc=4, frameon=False)
```

# Line Plots

```python
linear_data = np.array(range(8))
quadratic_data = linear_data ** 2

plt.figure()
plt.plot(linear_data, '-o', quadratic_data, '-o')
```

```python
plt.xlabel('Some Data')
plt.ylabel('Som Other data')
plt.title('Data')
plt.plot([3, 4, 5], '--r')
plt.legend(['Data 1', 'Data 2', 'Data 3'])
```

```python
plt.gca().fill_between(range(len(linear_data)),
                       linear_data,
                       quadratic_data,
                       alpha=0.25)
```

```python
plt.figure()

observation_data = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')

plt.plot(observation_data, linear_data, '-o',
         observation_data, quadratic_data, '-o')
```

```python
import pandas as pd

plt.figure()

observation_data = list(map(pd.to_datetime, observation_data))

plt.plot(observation_data, linear_data, '-o',
         observation_data, quadratic_data, '-o')
```

```python
x = plt.gca().xaxis

for item in x.get_ticklabels():
    item.set_rotation(45)

plt.subplots_adjust(bottom=.25)
```

```python
xrange = range(len(linear_data))

plt.figure()
plt.bar(xrange, linear_data, width=.3)
```

```python
new_list = []

for item in xrange:
    new_list.append(item + .3)

plt.bar(new_list, quadratic_data, width=.3, color='red')
```

```python
from random import randint

lin_err = [randint(0, 15) for x in range(len(linear_data))]
plt.bar(xrange, linear_data, width=.3, yerr=lin_err)
```

```python
plt.figure()

plt.bar(xrange, linear_data, width=.3, color='green')
plt.bar(xrange, quadratic_data, bottom=linear_data, width=.3, color='red')
```

```python
plt.figure()

plt.barh(xrange, linear_data, height=.3, color='g')
plt.barh(new_list, quadratic_data, height=.3, color='r')
```
